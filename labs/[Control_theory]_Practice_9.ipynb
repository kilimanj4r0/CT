# **Linear quadratic regulator (LQR)**
Consider the system:
$$ \dot{\mathbf{x}} = \mathbf{A}\mathbf{x}+\mathbf{B}\mathbf{u}$$
determines the matrix $\mathbf{K}$ of the optimal control law:
$$ \mathbf{u} = -\mathbf{K}\mathbf{x}$$

so as to minimize the performance index:
$$ J = \int^\infty_0(\mathbf{x}^T\mathbf{Q}\mathbf{x}+\mathbf{u}^T\mathbf{R}\mathbf{u})dt$$

where $\mathbf{Q}$ is a positive-definite (or positive-semidefinite) Hermitian or real symmetric matrix and $\mathbf{R}$ is a positive-definite Hermitian or real symmetric matrix.

**NOTE** that the
second term on the right-hand side of the perfomance index accounts for the expenditure of the energy of the control signals. The matrices $\mathbf{Q}$ and $\mathbf{R}$ determine the relative importance of the error and the expenditure of this energy. In this problem, we assume
that the control vector $\mathbf{u}$ is unconstrained.

Let's solve the optimization problem. Substitute the control law into the state space model:
$$  \dot{\mathbf{x}} = (\mathbf{A}-\mathbf{B}\mathbf{K})\mathbf{x}$$
And now we can substitute this law into the optimization problem:
$$ J = \int^\infty_0\mathbf{x}^T(\mathbf{Q}+\mathbf{K}^T\mathbf{R}\mathbf{K})\mathbf{x}dt$$

# **Riccati equation**

Let us set:
$$ \mathbf{x}^T(\mathbf{Q}+\mathbf{K}^T\mathbf{R}\mathbf{K})\mathbf{x} = -\frac{d}{dt}(\mathbf{x}^T\mathbf{P}\mathbf{x})$$
where $\mathbf{P}$ is a positive-definite Hermitian or real symmetric matrix.

So, we obtain following equation:
$$\mathbf{x}^T(\mathbf{Q}+\mathbf{K}^T\mathbf{R}\mathbf{K})\mathbf{x} = -\dot{\mathbf{x}}^T\mathbf{P}\mathbf{x} - \mathbf{x}^T\mathbf{P}\dot{\mathbf{x}} = -\mathbf{x}^T\left((\mathbf{A}-\mathbf{B}\mathbf{K})^T\mathbf{P}+\mathbf{P}(\mathbf{A}-\mathbf{B}\mathbf{K})\right)\mathbf{x}$$

Then:
$$-(\mathbf{Q}+\mathbf{K}^T\mathbf{R}\mathbf{K}) = (\mathbf{A}-\mathbf{B}\mathbf{K})^T\mathbf{P}+\mathbf{P}(\mathbf{A}-\mathbf{B}\mathbf{K})$$

It can be proved that if $\mathbf{A}-\mathbf{B}\mathbf{K}$ is a stable matrix, there exists a positive-definite matrix $\mathbf{P}$ that satisfies this equation.

Hence our procedure is to determine the elements of $\mathbf{P}$ from equation and
see if it is positive definite.

So, the perfomance index $J$ can be evaluated as:
$$ J = \int^\infty_0\mathbf{x}^T(\mathbf{Q}+\mathbf{K}^T\mathbf{R}\mathbf{K})\mathbf{x}dt = \int^\infty_0\left(-\frac{d}{dt}(\mathbf{x}^T\mathbf{P}\mathbf{x})\right)dt = -\mathbf{x}(\infty)^T\mathbf{P}\mathbf{x}(\infty)+\mathbf{x}(0)^T\mathbf{P}\mathbf{x}(0)$$

Since all eigenvalues of $\mathbf{A}-\mathbf{B}\mathbf{K}$ are assumed to have negative real parts, we have $\mathbf{x}(\infty) \to 0$
Therefore, we obtain:
$$ J = \mathbf{x}(0)^T\mathbf{P}\mathbf{x}(0)$$

Since $\mathbf{R}$ has been assumed to be a positive-definite Hermitian or real symmetric
matrix, we can write: $\mathbf{R} = \mathbf{T}^T\mathbf{T}$. Then:
$$
(\mathbf{A}^T-\mathbf{K}^T\mathbf{B}^T)\mathbf{P}+\mathbf{P}(\mathbf{A}-\mathbf{B}\mathbf{K})+\mathbf{Q}+\mathbf{K}^T\mathbf{T}^T\mathbf{T}\mathbf{K}=0
$$

This is equivalent to:
$$ \mathbf{A}^T\mathbf{P}+ \mathbf{P}\mathbf{A}+[\mathbf{T}\mathbf{K}-(\mathbf{T}^T)^{-1}\mathbf{B}^T\mathbf{P}][\mathbf{T}\mathbf{K}-(\mathbf{T}^T)^{-1}\mathbf{B}^T\mathbf{P}]-\mathbf{P}\mathbf{B}\mathbf{R}^{-1}\mathbf{B}^T\mathbf{P}+\mathbf{Q}=0$$

## **Bonus exercise**
> Write the derivation from the equation:
$$
(\mathbf{A}^T-\mathbf{K}^T\mathbf{B}^T)\mathbf{P}+\mathbf{P}(\mathbf{A}-\mathbf{B}\mathbf{K})+\mathbf{Q}+\mathbf{K}^T\mathbf{T}^T\mathbf{T}\mathbf{K}=0
$$
to:
$$ \mathbf{A}^T\mathbf{P}+ \mathbf{P}\mathbf{A}+[\mathbf{T}\mathbf{K}-(\mathbf{T}^T)^{-1}\mathbf{B}^T\mathbf{P}][\mathbf{T}\mathbf{K}-(\mathbf{T}^T)^{-1}\mathbf{B}^T\mathbf{P}]-\mathbf{P}\mathbf{B}\mathbf{R}^{-1}\mathbf{B}^T\mathbf{P}+\mathbf{Q}=0$$
with respect to: $\mathbf{R} = \mathbf{T}^T\mathbf{T}$


The minimization of $J$ with respect to $\mathbf{K}$ requires the minimization of:
$$
\mathbf{x}^T[\mathbf{T}\mathbf{K}-(\mathbf{T}^T)^{-1}\mathbf{B}^T\mathbf{P}][\mathbf{T}\mathbf{K}-(\mathbf{T}^T)^{-1}\mathbf{B}^T\mathbf{P}]\mathbf{x}
$$
Since this last expression is nonnegative, the minimum occurs when it is zero, or when
$$\mathbf{T}\mathbf{K} = (\mathbf{T}^T)^{-1}\mathbf{B}^T\mathbf{P}$$

Hence:
$$ \mathbf{K} = \mathbf{R}^{-1}\mathbf{B}^T\mathbf{P}$$

And the **Riccati equation** is:
$$ \mathbf{A}^T\mathbf{P}+ \mathbf{P}\mathbf{A}-\mathbf{P}\mathbf{B}\mathbf{R}^{-1}\mathbf{B}^T\mathbf{P}+\mathbf{Q}=0$$

## **Example**
Consider the following system:
$$\dot {\mathbf{x}} = \begin{bmatrix}0 & 1 \\ 0 & 0
\end{bmatrix}\mathbf{x}+
\begin{bmatrix} 0 \\ 1\end{bmatrix}\mathbf{u}$$

Let $\mathbf{Q} = \begin{bmatrix} 1 & 0 \\ 0 & 2
\end{bmatrix}$, $\mathbf{R}=\begin{bmatrix} 1\end{bmatrix}$ and $P = \begin{bmatrix} p_1 & p_2 \\ p_2 & p_3
\end{bmatrix}$

Let's solve Riccati equation:
$$ \mathbf{A}^T\mathbf{P}+ \mathbf{P}\mathbf{A}-\mathbf{P}\mathbf{B}\mathbf{R}^{-1}\mathbf{B}^T\mathbf{P}+\mathbf{Q}=0$$

This equation can be rewritten as:
$$
\begin{bmatrix}0 & 0 \\ 1 & 0
\end{bmatrix}\begin{bmatrix} p_1 & p_2 \\ p_2 & p_3
\end{bmatrix}+\begin{bmatrix} p_1 & p_2 \\ p_2 & p_3
\end{bmatrix}\begin{bmatrix}0 & 1 \\ 0 & 0
\end{bmatrix}-\begin{bmatrix} p_1 & p_2 \\ p_2 & p_3
\end{bmatrix}\begin{bmatrix} 0 \\ 1\end{bmatrix}\begin{bmatrix} 1\end{bmatrix}\begin{bmatrix} 0 & 1\end{bmatrix}\begin{bmatrix} p_1 & p_2 \\ p_2 & p_3
\end{bmatrix}+\begin{bmatrix} 1 & 0 \\ 0 & 2
\end{bmatrix}=\begin{bmatrix} 0 & 0 \\ 0 & 0
\end{bmatrix}
$$

This equation can be simplified to:
$$\begin{bmatrix} 0 & 0 \\ p_1 & p_2
\end{bmatrix}+\begin{bmatrix} 0 & p_1 \\ 0 & p_2
\end{bmatrix}-\begin{bmatrix} p_2^2 & p_2p_3 \\ p_2p_3 & p_3^2
\end{bmatrix}+\begin{bmatrix} 1 & 0 \\ 0 & 2
\end{bmatrix}=\begin{bmatrix} 0 & 0 \\ 0 & 0
\end{bmatrix}$$

From which we obtain the following three equations:
$$
\begin{cases}
1-p_2^2=0\\
p_1-p_2p_3=0\\
2-2p_2-p_3^2=0
\end{cases}
$$

The solution for these equations is: $p_1 = 2$, $p_2 = 1$ and $p_3 = 2$

So the $\mathbf{K}$ matrix is obtained as:
$$ \mathbf{K} = \mathbf{R}^{-1}\mathbf{B}^T\mathbf{P}=\begin{bmatrix} 1\end{bmatrix}\begin{bmatrix} 0 & 1\end{bmatrix}\begin{bmatrix} p_1 & p_2 \\ p_2 & p_3
\end{bmatrix}=\begin{bmatrix} p_2 & p_3\end{bmatrix}=\begin{bmatrix} 1 & 2\end{bmatrix}$$

Thus, the optimal control signal is
$$\mathbf{u}=-x_1-2x_2$$

```python
from matplotlib.pyplot import *
from scipy.integrate import odeint
from scipy.signal import place_poles
import numpy as np

def control(x, K):
  u = -np.dot(K,x)
  return u

def system_ode(x, t, A, B, K):
    u = control(x,K)
    dx = np.dot(A,x) + np.dot(B,u)
    return dx

t0 = 0 # Initial time 
tf = 10 # Final time
N = int(2E3) # Numbers of points in time span
t = np.linspace(t0, tf, N) # Create time span
x0= np.array([1,1])

A = np.array([[ 0, 1],
              [0,  0]])
B = np.array([[0],
              [1]])
Q = np.array([[1, 0],
              [0,  2]])
R = np.array([[1]])

from scipy.linalg import solve_continuous_are
P = solve_continuous_are(A,B,Q,R)
print("P:\n",P)

K = np.dot(np.dot(np.linalg.inv(R), np.transpose(B)),P)
print("K:\n",K)

x_sol = odeint(system_ode, x0, t, args=(A, B, K))

plot(t, x_sol[:,0], 'r', linewidth = 2., label = r'$x_1$ state')
plot(t, x_sol[:,1], 'b', linewidth = 2., label = r'$x_2$ state')
legend()
grid(color='black', linestyle='--', linewidth=1.0, alpha = 0.7)
grid(True)
ylabel(r'System state $x$')
xlabel(r'Time $t$')
show()
```

## **Comparison between pole placement and LQR techniques**

```python
def calculate_control(t,x_sol,K):
  u = np.zeros(t.shape)
  for i in range(len(t)):
    u[i]=control(x_sol[i],K)
  return u

#Pole placement
P_pp = np.array([-3, -5])
K_pp = place_poles(A, B, P_pp).gain_matrix

x_sol_pp = odeint(system_ode, x0, t, args=(A, B, K_pp))
u_pp = calculate_control(t,x_sol_pp,K_pp)


#LQR:
Q = np.array([[1, 0],
              [0,  1]])
R = np.array([[1]])

P_lqr = solve_continuous_are(A,B,Q,R)
K_lqr = np.dot(np.dot(np.linalg.inv(R), np.transpose(B)),P_lqr)

x_sol_lqr = odeint(system_ode, x0, t, args=(A, B, K_lqr))
u_lqr = calculate_control(t,x_sol_lqr,K_lqr)

plot(t, u_pp, 'r', linestyle = '--', linewidth = 2., label = r'Pole placement')
plot(t, u_lqr, 'r', linewidth = 2., label = r'LQR')
legend()
grid(color='black', linestyle='--', linewidth=1.0, alpha = 0.7)
grid(True)
ylabel(r'Control signal $u$')
xlabel(r'Time')
show()

plot(t, x_sol_pp[:,0], 'r', linestyle='--', linewidth = 2., label = r'$x_1$ state (pole placement)')
plot(t, x_sol_lqr[:,0], 'r', linewidth = 2., label = r'$x_1$ state (LQR)')
plot(t, x_sol_pp[:,1], 'b', linestyle='--', linewidth = 2., label = r'$x_2$ state (pole placement)')
plot(t, x_sol_lqr[:,1], 'b', linewidth = 2., label = r'$x_2$ state (LQR)')
legend()
grid(color='black', linestyle='--', linewidth=1.0, alpha = 0.7)
grid(True)
ylabel(r'System state $x$')
xlabel(r'Time $t$')
show()
```
